```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("/content/spotify dataset.csv") #Load Dataset
df.head()
```

```python
df.info() #Basic Info
df.describe()
```

```python
#Converting release date
df['track_album_release_date'] = pd.to_datetime(
    df['track_album_release_date'], errors='coerce'
)
```

```python
df.drop_duplicates(subset='track_id', inplace=True) #Remove duplicates
```

```python
plt.figure() #Popularity Distribution
sns.histplot(df['track_popularity'], bins=30)
plt.title("Track Popularity Distribution")
plt.show()
```

```python
plt.figure() #Genre Distribution
df['playlist_genre'].value_counts().plot(kind='bar')
plt.title("Songs per Genre")
plt.show()
```

```python
sns.pairplot(df[['danceability','energy','valence','tempo']]) #Feature Relationships
plt.show()
```

```python
df.isnull().sum() #Handle missing values

# Fill numeric columns with median
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
plt.figure(figsize=(12,8)) #Correlation Matrix
corr = df[num_cols].corr()
sns.heatmap(corr, cmap='coolwarm')
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
#Select features for clustering
features = [
    'danceability','energy','loudness','speechiness',
    'acousticness','instrumentalness','liveness',
    'valence','tempo'
]

X = df[features]
```

```python
#Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#K-Means Clustering
kmeans = KMeans(n_clusters=5, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
#Plot clusters
plt.figure()
sns.scatterplot(
    x=df['danceability'],
    y=df['energy'],
    hue=df['cluster'],
    palette='tab10'
)
plt.title("Song Clusters")
plt.show()
```

```python
#Clusters by genre
plt.figure()
sns.countplot(x='cluster', hue='playlist_genre', data=df)
plt.title("Clusters by Genre")
plt.show()
```

```python
#Compute similarity matrix
similarity = cosine_similarity(X_scaled)
```

```python
#Recommendation function
def recommend_song(song_name, df, similarity, n=5):
    idx = df[df['track_name'] == song_name].index[0]
    scores = list(enumerate(similarity[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    scores = scores[1:n+1]

    song_indices = [i[0] for i in scores]
    return df[['track_name','track_artist']].iloc[song_indices]
```

```python
#Example
recommend_song("Memories - Dillon Francis Remix", df, similarity)
```
